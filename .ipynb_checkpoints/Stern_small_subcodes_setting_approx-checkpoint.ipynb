# Same as other script, but we assume amortized lists (to make search faster)
# Also, we use approximation for binomials

```python
reset()

import warnings
warnings.filterwarnings("ignore")

def h2(u):
    return -u*log(N(u),2)-(1-u)*log(1-N(u),2)

def approx_log_binomial(y, x):
    return 0.5*log(N(y/(2*3.14159265358*x*(y-x))),2)+y*h2(x/y)
    
# min distance of ldpc codes from the ensemble
def ldpc_min_dist(n, k, nu):
    d = 1
    val = -1
    while val < 0:
        d+= 1
        val = log(binomial(n,d),2)-(n-k)+(n-k)*log(1+(1-2*nu)^d,2)
    return d

#find ell2 value yielding amortized lists
#NOTE: I wrote my own function, cause the build-in one gives fucking problems
def compute_ell2(log_L, ell2_max, k, ell1, z, p2):
    
    #We find ell2_th using bisection
    left_ell2 = 1
    right_ell2 = ell2_max
    val_left = log_L - approx_log_binomial(k + ell1 + left_ell2-z, p2) #this should be negative
    val_right = log_L - approx_log_binomial(k + ell1 + right_ell2-z, p2) #this should be positive
    
    #if both have the same sign, a solution cannot be found
    if (val_left*val_right)>0:
        return -1
    
    #use bisection to find a solution
    while round(left_ell2) != round(right_ell2):
        
        val_left = log_L - approx_log_binomial(k + ell1 + left_ell2-z, p2) #this should be negative
        val_right = log_L - approx_log_binomial(k + ell1 + right_ell2-z, p2) #this should be positive
        
        ell2_middle = (left_ell2+right_ell2)/2
        val_middle = log_L - approx_log_binomial(k + ell1 + ell2_middle-z, p2) #this, we don't know
        
  #      print(val_left, val_middle, val_right)
  #      print(round(left_ell2), N(ell2_middle), round(right_ell2))
        
        #bisect again
        if (val_left*val_middle)>0: #both are positive, set val_left = val_middle
            left_ell2 = ell2_middle
        else:
            right_ell2 = ell2_middle
   #     print(" ")
   # print(" ------- ")
    
    return floor(left_ell2)
```

# Cost of new algorithm

```python
#cost of new alforithm
def cost_sparse_stern(n,k,nu,w):
    
    ell1_max = 100
    ell2_max = 100 #this can be high, it does not suffer much from brute force search
    p1_max = 10
    p2_max = 10
    
    min_cost = 1000000000000000000000000000
    for ell1 in range(1,min(ell1_max, n-k)+1):

        z = floor((1-(1-nu)^ell1)*n)
        if z<=k:
            for p1 in range(1,min(floor(z/2), p1_max)+1):
                log_L1 = approx_log_binomial(floor(z/2),p1)
                log_L2 = approx_log_binomial(ceil(z/2),p1)
                log_L = log_L1 + log_L2 -ell1

                #check if this is already not optimal            
                if (max(log_L,log_L2) + approx_log_binomial(n,w)-(approx_log_binomial(ceil(z/2),p1) +approx_log_binomial(floor(z/2),p1) +approx_log_binomial(n-z,w-2*p1)))<min_cost:
                    for p2 in range(1,min(p2_max, floor(z/2))):
                    
                        #search for ell2 yielding amortized lists
                        ell2_th = compute_ell2(log_L, min(ell2_max, n-k-ell1), k, ell1, z, p2)
                        if ell2_th>0:
                            #checking ell2 and ell2+1
                            ell2_th_2 = round(log_L)
                            for ell2 in range(min(ell2_th, ell2_th_2)-2, max(ell2_th, ell2_th_2)-2+2):
                                m = k+ell1+ell2-z
                                log_M = approx_log_binomial(m, p2)
                                log_K = log_M + log_L - ell2

                                log_pr = approx_log_binomial(ceil(z/2),p1) + approx_log_binomial(floor(z/2),p1)+ approx_log_binomial(m, p2) + approx_log_binomial(n-(k+ell1+ell2),w-2*p1-p2) - approx_log_binomial(n,w)
                                cost = log(n^3+n*(2^log_L1+2^log_L2+2^log_L+2^log_M+2^log_K),2)-log_pr

                                if cost < min_cost:
                                    min_cost = cost
                                    best_params = [ell1, ell2, p1, p2]
                                    quantities = [log_L1, log_L, log_M, log_K]
#                                        print(ell1, z, ell2, p1, p2, m)
#                                        print(N(min_cost), N(log(L1, 2)), N(log(L, 2)), N(log(M, 2)), N(log(K, 2)))
#                                        print("-----")

    #Sanity check on paramters
    if best_params[0] == ell1_max:
        print("!!! SPARSE STERN: ell1_max too low !!!")
    if best_params[1] == ell2_max:
        print("!!! SPARSE STERN: ell2_max too low !!!")
    if best_params[2] == p1_max:
        print("!!! SPARSE STERN: p1_max too low !!!")
    if best_params[3] == p2_max:
        print("!!! SPARSE STERN: p2_max too low !!!")
        
    return quantities, best_params, min_cost
```

# Cost of traditional Stern

```python
#cost of old Stern
def cost_stern(n,k,w):
    
    ell_max = 100
    p_max = 20
    
    min_cost = 1000000000000000000000000000
    for ell in range(1,min(ell_max, n-k)+1):
        z = k+ell
        for p in range(1,min(floor(z/2), p_max)+1):
            L1 = binomial(floor(z/2),p)
            L2 = binomial(ceil(z/2),p)
            L = L1*L2*2^(-ell)
            

            pr = binomial(ceil(z/2),p)*binomial(floor(z/2),p)*binomial(n-k-ell,w-2*p)/binomial(n,w)
            cost = log(n^3+n*(L1+L2+L),2)-log(pr,2)

            if cost < min_cost:
                min_cost = cost
                best_params = [ell, p]


    #Sanity check on paramters
    if best_params[0] == ell_max:
        print("!!! STERN: ell_max too low !!!")
    if best_params[1] == p_max:
        print("!!! STERN: p_max too low !!!")
        
    return best_params, min_cost
```

# Get how parameters change

We fix density (as a function of $n$) and code rate $R$, then test several values of $n$; for each value of $n$:

- we consider minimum distances of LDPC code and RND code with same parameters

- we consider the costs of both SparseStern and Stern on LDPC

- we consider also the costs of Stern on RND code

```python
def density(n):
    return N(log(n)^2/n)
#    return 0.05

#params for simulation
R = 0.8
n_min = 500
n_max = 10000
n_step = 500

ell1_vals = []
ell2_vals = []
p1_vals = []
p2_vals = []

sparse_stern_costs = []
old_stern_costs = []
RND_stern_costs = []
    
for n in range(n_min, n_max+1, n_step):
    
    #get minimum distances
    k = floor(R*n)
    nu = density(n)
    d_ldpc = ldpc_min_dist(n, k, nu)
    d_rnd = ldpc_min_dist(n, k, 0.5)
        
    print("n = ",n,", d_ldpc = ",d_ldpc,", d_RND = ",d_rnd)
    print("--> k = ",k)
    print("--> nu = ",nu)
    print("--> E[wt(h_i)] = ",N(nu*(n-k)))
    print(" ")
    
    #Sparse Stern
    sparse_quantities, sparse_stern_params, sparse_stern_cost = cost_sparse_stern(n, k, nu, d_ldpc)
    print("SparseStern cost = ",N(sparse_stern_cost),", coeff. = ",N(sparse_stern_cost/n))
    print("---> ell1 = ",sparse_stern_params[0])
    print("---> ell2 = ",sparse_stern_params[1])
    print("---> p1 = ",sparse_stern_params[2])
    print("---> p2 = ",sparse_stern_params[3])
    print("---> L1_prime = ",sparse_quantities[0])
    print("---> L = ",sparse_quantities[1])
    print("---> L2 = ",sparse_quantities[2])
    print("---> M = ",sparse_quantities[3])
    
    #Old Stern
    old_stern_params, old_stern_cost = cost_stern(n, k, d_ldpc)
    print("Stern cost = ",N(old_stern_cost),", coeff. = ",N(old_stern_cost/n))
    
    #Old Stern, RND code
    if d_ldpc != d_rnd:
        RND_old_stern_params, RND_old_stern_cost = cost_stern(n, k, d_rnd)
    else:
        RND_old_stern_cost = old_stern_cost    
    print("RND Stern cost = ",N(RND_old_stern_cost),", coeff. = ",N(RND_old_stern_cost/n))
    
    
    print("=================================")
    
    #Append optimal parameters values
    ell1_vals.append((n, sparse_stern_params[0]))
    ell2_vals.append((n, sparse_stern_params[1]))
    p1_vals.append((n, sparse_stern_params[2]))
    p2_vals.append((n, sparse_stern_params[3]))
    
    #Append costs
    sparse_stern_costs.append((n, sparse_stern_cost/n))
    old_stern_costs.append((n, old_stern_cost/n))
    RND_stern_costs.append((n, RND_old_stern_cost/n))
```
